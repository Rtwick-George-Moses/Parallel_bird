```python
from numpy import asarray
from numpy import savetxt
from numpy import loadtxt
import sys
import time
import numpy as np
sys.setrecursionlimit(10**6)

time_holder = asarray([0,0,0])
generations = asarray([1,1,1])
# save to csv file
savetxt('time.csv', time_holder, delimiter=',')
savetxt('generations.csv', generations, delimiter=',')

prev_time_hold = loadtxt('time.csv', delimiter=',')
prev_gen_hold = loadtxt('generations.csv', delimiter=',')
remaing_iterations = 0
vals = asarray([5,3,2])
def control_thread(total_iterations):
    global prev_gen_hold
    generation_current = loadtxt('generations.csv', delimiter=',')
    time_current = loadtxt('time.csv', delimiter=',')
    remaing_iterations = total_iterations
    
    if remaing_iterations <= 0:
        end_time = asarray([-1,-1,-1])
        savetxt('generations.csv', end_time, delimiter=',')
        
    if time_current.all() == prev_time_hold.all():
        time.sleep(3)
        control_thread(remaing_iterations)
        
        
    if time_current.all() != prev_time_hold.all():
        print("appending the vals----|||||||")
        prev_gen_hold = generation_current
        print("updating")
        for i in range(len(time_current)):
            generation_current[time_current.argmax()] = generation_current[time_current.argmax()] + vals[i]
            time_current[time_current.argmax()] = 0
        savetxt('time.csv', time_current, delimiter=',')
        savetxt('generations.csv', generation_current, delimiter=',')
        print(generation_current)
        print(prev_time_hold)
        control_thread(remaing_iterations - np.sum(vals))
            
        
```

```python
def thread1():
    from model1 import flappy1
    flappy1.run()
def thread2():
    from model2 import flappy2
    flappy2.run()
def thread3():
    from model3 import flappy3
    flappy3.run()
```

```python
import threading
  
# creating processes
p1 = threading.Thread(target=thread1, args=())
p2 = threading.Thread(target=thread2, args=())
p3 = threading.Thread(target=thread3, args=())
p4 = threading.Thread(target=control_thread, args=(50,))


  
p1.start()
p2.start()
p3.start()
p4.start()
    
p1.join()
p2.join()
p3.join()
p4.join()
print("Done!")
```
